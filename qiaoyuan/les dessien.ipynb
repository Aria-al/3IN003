```python
import matplotlib.pyplot as plt
```

```python


# 数据
x = [250, 500, 750, 1000, 1250, 1500, 1750, 2000]
y = [0.00200, 0.06300, 0.42100, 1.87400, 6.03600, 16.21600, 38.43000, 82.48500]

# 绘图
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o')
plt.xlabel("S")
plt.ylabel("Temps(s)")
plt.title("Plot of Times vs S, Algo1, k=8, d=3")
plt.grid(True)
plt.show()
```
